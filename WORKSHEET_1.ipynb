```python
import numpy as np
x = np.array([[1, 2],
              [3, 5]])

y = np.array([[5, 6],
              [7, 8]])

v = np.array([9, 10])
w = np.array([11, 12])

result_vw = np.dot(v, w)
print("Dot product of v and w:", result_vw)

result_xv = np.dot(x, v)
print("Dot product of x and v:", result_xv)

result_xy = np.dot(x, y)
print("Dot product of x and y:")
print(result_xy)
```

```python
import numpy as np

x = np.array([[1,2],[3,5]])
y = np.array([[5,6],[7,8]])
v = np.array([9,10])
w = np.array([11,12])

concat_xy_row = np.vstack((x, y))
print("Concatenate x and y along rows:\n", concat_xy_row)

concat_vw_col = np.hstack((v, w))
print("Concatenate v and w along columns:\n", concat_vw_col)
```

```python
import numpy as np

x = np.array([[1,2],[3,5]])
v = np.array([9,10])

try:
    concat_xv = np.concatenate((x, v))
    print("Concatenation successful:\n", concat_xv)
except ValueError as e:
    print("Error concatenating x and v:", e)
```

The error occurs because 'x' is a 2-dimensional array (shape (2, 2)) and 'v' is a 1-dimensional array (shape (2,)). • Solve the following system of Linear equation using Inverse Methods.

2x − 3y + z = −1
x − y + 2z = −3
3x + y − z = 9

{Hint: First use Numpy array to represent the equation in Matrix form. Then Solve for: AX = B}
• Now: solve the above equation using np.linalg.inv function.{Explore more about ”linalg” function

```python
import numpy as np
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])
A1 = np.linalg.inv(A)
I = A.dot(A1)
I = np.round(I)
print("A*A^-1=I:\n",I)
```

```python
import numpy as np
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])
AB = A.dot(B)
BA = B.dot(A)
print("AB:\n",AB)
print("BA:\n",BA)
print("AB != BA")
```

```python
import numpy as np
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])
left_side = AB.T
right_side = B.T.dot(A.T)
print(("(AB)^T:\n"),left_side)
print("B^T A^T:\n",right_side)
print("(AB)^T = B^T A^T")
```
