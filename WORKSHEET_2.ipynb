Complete all the following Task:

1. Load the provided dataset and import in pandas DataFrame.

3. For the column with missing values fill the values using various techniques we discussed above. Try
to explain why did you select the particular methods for particular column.
4. Check for any duplicate values present in Dataset and do necessary to manage the duplicate items.
{Hint: dataset.duplicated.sum()}


```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Worksheet/medical_students_dataset.csv')

# Display first 5 rows
print(df.head())

# Display shape of dataset
print("Shape:", df.shape)

# Display info about dataset
print(df.info())
```

```python
import pandas as pd

# Assuming df is already loaded
# Check info of the DataFrame
print(df.info())

# Check total number of null values in each column
null_counts = df.isnull().sum()
print(null_counts)

# list only columns that have missing values
missing_cols = null_counts[null_counts > 0]
print("Columns with missing values:")
```

```python
for col in df.columns:
    if df[col].isnull().sum() > 0:
        if df[col].dtype == 'object':
            df[col] = df[col].fillna(df[col].mode()[0])
        else:
            df[col] = df[col].fillna(df[col].mean())

print("Null values after imputation:")
print(df.isnull().sum())
```

```python
import pandas as pd

duplicates = df.duplicated().sum()
print("Total duplicate rows:", duplicates)

df_cleaned = df.drop_duplicates()
print("Duplicates after removal:", df_cleaned.duplicated().sum())
```

Problem - 1:
Create a DataFrame that is subsetted for the columns ’Name’, ’Pclass’, ’Sex’, ’Age’, ’Fare’, and ’Survived’.
Retain only those rows where ’Pclass’ is equal to 1, representing first-class passengers. What is the mean,
median, maximum value, and minimum value of the ’Fare’ column?

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Worksheet/Titanic-Dataset.csv')

subset_cols = ['Name', 'Pclass', 'Sex', 'Age', 'Fare', 'Survived']
titanic_subset = df[subset_cols]

first_class = titanic_subset[titanic_subset['Pclass'] == 1]

print("First-class passengers subset:")
print(first_class.head())

fare_mean = first_class['Fare'].mean()
fare_median = first_class['Fare'].median()
fare_max = first_class['Fare'].max()
fare_min = first_class['Fare'].min()

print("\nStatistics of 'Fare' for first-class passengers:")
print(f"Mean Fare: {fare_mean}")
print(f"Median Fare: {fare_median}")
print(f"Maximum Fare: {fare_max}")
print(f"Minimum Fare: {fare_min}")
```

Problem - 4:

Compare the mean survival rates (’Survived’) for the different groups in the ’Sex’ column. Draw a visual-
ization to show how the survival distributions vary by gender.

```python
import pandas as pd

import matplotlib.pyplot as plt

survival_by_sex = df.groupby('Sex')['Survived'].mean()
print("Mean survival rate by Sex:")
print(survival_by_sex)

survival_by_sex.plot(kind='bar', color=['skyblue', 'lightgreen'])
plt.title('Survival Rate by Gender')
plt.ylabel('Mean Survival Rate')
plt.ylim(0, 1)
plt.show()
```
